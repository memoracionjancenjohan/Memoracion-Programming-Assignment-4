```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('board2.csv')
df
```

```python
df.describe()
```

```python
Instru = df.loc[(df['Track'] == 'Instrumentation') & (df['Electronics'] > 70) & (df['Hometown'] == 'Luzon')][['Name', 'GEAS', 'Electronics']]
Instru
```

```python
df['Average'] = df[['Math', 'Electronics', 'GEAS']].mean(axis = 1)
Mindy = df.loc[(df['Average'] >= 55) & (df['Hometown'] == 'Mindanao') & (df['Gender'] == 'Female')][['Name', 'Track', 'Electronics', 'Average']]
Mindy
```

```python
#Average by Track
avg_by_track = df.groupby("Track")["Average"].mean()
plt.figure(figsize=(6,4))
avg_by_track.plot(kind="bar")
plt.title("Average Grades by Track")
plt.xlabel("Track")
plt.ylabel("Average Grade")
plt.show()
```

```python
#Average by Gender
avg_by_gender = df.groupby("Gender")["Average"].mean()
plt.figure(figsize=(4,4))
avg_by_gender.plot(kind="bar", color="orange")
plt.title("Average Grades by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Grade")
plt.show()
```

```python
#Average by Hometown
avg_by_home = df.groupby("Hometown")["Average"].mean()
plt.figure(figsize=(6,4))
avg_by_home.plot(kind="bar", color="green")
plt.title("Average Grades by Hometown")
plt.xlabel("Hometown")
plt.ylabel("Average Grade")
plt.show()
```
